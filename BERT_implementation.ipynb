```python
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
preprocessor = hub.KerasLayer(
    "https://kaggle.com/models/tensorflow/bert/TensorFlow2/en-uncased-preprocess/3")

encoder = hub.KerasLayer(
    "https://kaggle.com/models/tensorflow/bert/TensorFlow2/en-uncased-l-12-h-768-a-12/3")
```

```python
 text = ['what a great day','sun rises from the east']
```

## Preprocessing Stage

```python
text_preprocessed = preprocessor(text)
```

```python
text_preprocessed.keys()
```

```python
text_preprocessed['input_mask']
```

##### In the first sentence we can see 1,1,1,1,1,1, 6 1s because of CLS,what,a,great,day,SEP. Rest zeroes are padding, total is 128

```python
text_preprocessed['input_word_ids']
```

##### These are unqiue ids assigned to each word in the two sentences. See, CLS and SEP have same ids in both sentences.

## Getting Vector Embeddings

```python
bert_results  = encoder(text_preprocessed)
```

```python
bert_results.keys()
```

```python
bert_results['pooled_output']
```

##### pooled output is embedding for one full sentence (768 size vector for each sentence)

```python
bert_results['sequence_output']
```

##### sequence coutput is embedding for each word in every sentence

```python
len(bert_results['encoder_outputs'])
```

```python
bert_results['encoder_outputs']
```

#### output of each transformer encoder layer in the BERT model. Here we have BERT base model so 12 encoder layers showing.

```python
bert_results['encoder_outputs'][-1] == bert_results['sequence_output']
```

##### the last embedding of encoder_output is same as sequence output
